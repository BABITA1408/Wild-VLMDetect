```python
import os
import cv2
import numpy as np
from ultralytics import YOLO
from PIL import Image

# Ensure crop output folder exists
os.makedirs("/content/crops", exist_ok=True)

# Load trained model
model = YOLO("/content/runs/detect/train/weights/best.pt")

# Run inference
results = model(source="injured_dog_1.mp4", stream=True)

# Manual frame counter
frame_id = 0

# Loop over results (frames)
for r in results:
    frame = r.orig_img.copy()
    boxes = r.boxes.xyxy.cpu().numpy()
    classes = r.boxes.cls.cpu().numpy()

    for i, box in enumerate(boxes):
        label = model.names[int(classes[i])]
        if label in ["dog", "injured"]:
            x1, y1, x2, y2 = map(int, box)
            cropped = frame[y1:y2, x1:x2]

            # Use frame_id instead of r.frame
            filename = f"/content/crops/{label}_frame{frame_id}_box{i}.jpg"
            cv2.imwrite(filename, cropped)
            print(f"✅ Saved: {filename}")

    frame_id += 1
```

```python
from transformers import CLIPProcessor, CLIPModel

# Load CLIP model and processor
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Generate prompts from all YOLO classes
yolo_classes = list(model.names.values())  # from your already loaded YOLO model

# Create healthy/injured prompts for each class (except 'injured' itself)
clip_prompts = [f"a healthy {cls}" for cls in yolo_classes if cls != "injured"]
clip_prompts += [f"an injured {cls}" for cls in yolo_classes if cls != "injured"]
clip_prompts += ["a wild animal", "a domestic animal", "an unknown object"]
```

```python
from PIL import Image

def run_clip_on_crop(image_path, text_prompts):
    image = Image.open(image_path).convert("RGB")
    inputs = clip_processor(text=text_prompts, images=image, return_tensors="pt", padding=True)
    outputs = clip_model(**inputs)

    probs = outputs.logits_per_image.softmax(dim=1)[0]  # Normalize to get probabilities
    result = list(zip(text_prompts, probs.detach().numpy()))

    result.sort(key=lambda x: -x[1])  # Sort by confidence
    return result
```

```python
crop_dir = "/content/crops"

for filename in os.listdir(crop_dir):
    if filename.endswith(".jpg"):
        path = os.path.join(crop_dir, filename)
        predictions = run_clip_on_crop(path, clip_prompts)

        print(f"\n🔍 CLIP predictions for {filename}:")
        for text, score in predictions[:3]:  # Top 3 results
            print(f"  {text}: {score:.3f}")
```

```python
import csv

# Output CSV path
csv_path = "/content/clip_predictions.csv"

# Prepare CSV file
with open(csv_path, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["filename", "top_prediction", "confidence"])

    # Process crops
    for filename in os.listdir(crop_dir):
        if filename.endswith(".jpg"):
            path = os.path.join(crop_dir, filename)
            predictions = run_clip_on_crop(path, clip_prompts)
            top_text, top_score = predictions[0]

            # Write to CSV
            writer.writerow([filename, top_text, round(top_score, 4)])
            print(f"📄 Saved: {filename} → {top_text} ({top_score:.3f})")

print(f"\n✅ All predictions saved to: {csv_path}")
```

```python

from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can use "gpt2-medium", "gpt2-large", etc. for more powerful models
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Ensure the tokenizer is set to pad tokens correctly
tokenizer.pad_token = tokenizer.eos_token

# Function to generate text using GPT-2
def generate_text(prompt):
    inputs = tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True)
    outputs = model.generate(inputs['input_ids'], max_length=100, num_return_sequences=1, no_repeat_ngram_size=2, top_p=0.95, top_k=60)

    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

# Example usage: generate a caption for a cropped image (simulated as a prompt)
prompt = "An injured dog"  # You can modify the prompt based on the object detected by YOLO
caption = generate_text(prompt)
print("Generated Caption:", caption)
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import torch

# Load GPT-2 model and tokenizer
gpt2_model = GPT2LMHeadModel.from_pretrained("gpt2")
gpt2_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
gpt2_tokenizer.pad_token = gpt2_tokenizer.eos_token

# Load BLIP model for image captioning
blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Function to generate text using GPT-2
def generate_text(prompt):
    inputs = gpt2_tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True)
    outputs = gpt2_model.generate(inputs['input_ids'], max_length=100, num_return_sequences=1, no_repeat_ngram_size=2, top_p=0.95, top_k=60)
    generated_text = gpt2_tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

# Function to generate captions using BLIP
def generate_caption(image_path):
    image = Image.open(image_path).convert("RGB")
    inputs = blip_processor(images=image, return_tensors="pt")
    out = blip_model.generate(**inputs)
    caption = blip_processor.decode(out[0], skip_special_tokens=True)
    return caption

# Example of integrating BLIP and GPT-2
def detect_and_generate_description(image_path):
    # Generate caption using BLIP
    caption = generate_caption(image_path)

    # Classify the object detected by YOLO (this part comes from YOLO + CLIP predictions)
    # For this example, let's say CLIP predicted "injured cow"
    object_description = "An healthy cow"

    # Combine BLIP caption and object description into a prompt for GPT-2
    prompt = f"Description of an object: {object_description}. The scene is: {caption}. Describe this object in more detail."

    # Generate detailed text using GPT-2
    detailed_description = generate_text(prompt)
    return detailed_description

# Example usage
image_path = "/content/crops/cow_frame0_box0.jpg"  # Replace with your image path
generated_description = detect_and_generate_description(image_path)
print("Generated Description:", generated_description)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

true_labels = ['Injured Cow', 'Healthy Cow', 'Injured Dog', 'Injured Dog', 'Injured Cat']
pred_labels = ['Injured Cow', 'Healthy Cow', 'Injured Dog', 'Injured Dog', 'Healthy Cat']

labels = ['Injured Cow', 'Healthy Cow', 'Injured Dog', 'Injured Cat']
cm = confusion_matrix(true_labels, pred_labels, labels=labels)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
